## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana korelasi antara kondisi cuaca dengan jumlah sewa sepeda per harinya ?




- Dari hasil dibawah ini bisa dilihat bahwa jumlah penyewa sepeda banyak menyewa ketika cuaca dalam keadaan cerah (sebanyak 8714 sepeda dalam tiap harinya), dan paling sedikit menyewa dalam cuaca hujan ringan (sebanyak 22 sepeda saja dalam tiap harinya)

```python
day_df.groupby(by="weathersit").agg({
    # "instant": "nunique",
    "cnt": ["max", "min", "mean", "std"]
})
```

- Kemudian kita cek untuk total di setiap cuaca cerah ada berapa banyak sepeda yang disewakan, untuk membuktikan hasil analisis diatas, dan dibawah in hasil membuktikan dengan didapatkan total jumlah penyewa sepeda di cuaca cerah sebanyak 2257952 sepeda, dan penyewa sepeda di cuaca hujan ringan sebanyak 37869, jadi dapat disimpulkan dalam setiap harinya penyewa sepeda lebih cenderung melakukan penyewaan pada saat cuaca cerah.

```python
sum_order_items_df = day_df.groupby("weathersit").cnt.sum().sort_values(ascending=False).reset_index()
print(sum_order_items_df)
```

- Selanjutnya kita tampilkan dalam bentuk histogram

```python
fig, ax = plt.subplots(figsize=(12, 6))

colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

# Mapping cuaca dengan label
weather_labels = {
    1: 'Cerah',
    2: 'Berawan',
    3: 'Hujan Ringan',
}

# Mengganti nilai weathersit dengan label
sum_order_items_df['weathersit_label'] = sum_order_items_df['weathersit'].map(weather_labels)

# Merubah range pada sumbu y
y_range = {
    'Cerah': 2257952,
    'Berawan': 996858,
    'Hujan Ringan': 37869,
}

sns.barplot(x="weathersit_label", y="cnt", data=sum_order_items_df.head(5), palette=colors, ax=ax)
ax.set_ylabel("Jumlah Total Penyewa Sepeda")
ax.set_xlabel("Cuaca")
ax.set_title("Korelasi Kondisi Cuaca dengan Jumlah Sewa Sepeda", loc="center", fontsize=15)
ax.tick_params(axis='y', labelsize=12)

# Mengatur range pada sumbu y
ax.set_yticks([0, 500000, 1000000, 1500000, 2000000, 2500000])
ax.set_yticklabels(['0', '500,000', '1,000,000', '1,500,000', '2,000,000', '2,500,000'])

plt.show()
```

### Pertanyaan 2: Bagaimana korelasi antara kondisi cuaca dengan jumlah sewa sepeda per jam nya ?

- Dari hasil yang didapatkan bisa dilihat dibawah ini bahwa jumlah penyewa sepeda banyak menyewa ketika cuaca cerah dalam keadaan cerah (sebanyak 977 sepeda per jam nya), dan paling sedikit menyewa dalam cuaca badai petir (sebanyak 23 sepeda per jam nya)

```python
#korelasi per jam

hour_df.groupby(by="weathersit").agg({
    # "instant" : "nunique",
    "cnt":["max", "min", "mean", "std"]
})
```

- Kemudian kita cek untuk total di setiap cuaca cerah ada berapa banyak sepeda yang disewakan, untuk membuktikan hasil analisis diatas, dan dibawah ini hasil membuktikan dengan didapatkan total jumlah penyewa sepeda di cuaca cerah sebanyak 2338173 sepeda di setiap jam nya, dan penyewa sepeda di cuaca hujan lebat sebanyak 223, jadi dapat disimpulkan dalam setiap jam nya penyewa sepeda lebih cenderung melakukan penyewaan pada saat cuaca cerah.

```python
sum_order_items_df = hour_df.groupby("weathersit").cnt.sum().sort_values(ascending=False).reset_index()
print(sum_order_items_df)
```

- Selanjutnya kita tampilkan dalam bentuk histogram

```python
fig, ax = plt.subplots(figsize=(12, 6))

colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

# Mapping cuaca dengan label
weather_labels = {
    1: 'Cerah',
    2: 'Berawan',
    3: 'Hujan Ringan',
    4: 'Hujan Lebat'
}

# Mengganti nilai weathersit dengan label
sum_order_items_df['weathersit_label'] = sum_order_items_df['weathersit'].map(weather_labels)

# Merubah range pada sumbu y
y_range = {
    'Cerah': 2338173,
    'Berawan': 795952,
    'Hujan Ringan': 158331,
    'Hujan Lebat' : 223
}

sns.barplot(x="weathersit_label", y="cnt", data=sum_order_items_df.head(5), palette=colors, ax=ax)
ax.set_ylabel("Jumlah Total Penyewa Sepeda")
ax.set_xlabel("Cuaca")
ax.set_title("Korelasi Kondisi Cuaca dengan Jumlah Sewa Sepeda", loc="center", fontsize=15)
ax.tick_params(axis='y', labelsize=12)

# Mengatur range pada sumbu y
ax.set_yticks([0, 500000, 1000000, 1500000, 2000000, 2500000])
ax.set_yticklabels(['0', '500,000', '1,000,000', '1,500,000', '2,000,000', '2,500,000'])

plt.show()
```
